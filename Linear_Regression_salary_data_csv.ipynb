```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1.1,1.3,1.5,2,2.2,2.9,3,3.2,3.2,3.7,3.9,4,4,4.1,4.5,4.9,5.1,5.3,5.9,6,6.8,7.1,7.9,8.2,8.7,9,9.5,9.6,10.3,10.5])
y=np.array([39343,46205,37731,43525,39891,56642,60150,54445,64445,57189,63218,55794,56957,57081,61111,67938,66029,83088,81363,93940,91738,98273,101302,113812,109431,105582,116969,112635,122391,121872])          
plt.scatter(x,y,color='red')
plt.plot(x,y,color='green')
plt.title('x vs y')
plt.xlabel('x data')
plt.ylabel('y data')
plt.show()
```

```python
def predict(m,c,x):
    arr=[]
    for i in range(len(x)):
        a=m*x[i]+c
        arr.append(a)
    return arr
```

```python
def error(y,yh):
    ln=len(y)
    arr=[]
    for i in range(ln):
        a=(y[i]-yh[i])**2
        arr.append(a)
    avg_e=sum(arr)/ln
    return avg_e
```

```python
def dm_fun(x,y,yh):
    arr=[]
    for i in range(len(x)):
        a=(y[i]-yh[i])*x[i]
        arr.append(a)
    dm=(-2/len(x))*sum(arr)
    return dm
```

```python
def dc_fun(x,y,yh):
    arr=[]
    for i in range(len(x)):
        a=(y[i]-yh[i])
        arr.append(a)
    dc=(-2/len(x))*sum(arr)
    return dc
```

```python
m=0
c=0
l=0.0001
big_n=5000
err=[]
for i in range(big_n):
    yh=predict(m,c,x)
    e=error(y,yh)
    err.append(e)
    dm=dm_fun(x,y,yh)
    dc=dc_fun(x,y,yh)
    m=m-l*dm
    c=c-l*dc
print(m, c)
```

```python
plt.plot(err)
```

```python
yhn=[(m*xi+c) for xi in x]
print(yhn)
```
