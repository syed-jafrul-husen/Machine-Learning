```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70]
y = [90,115,130,157,170,210,250]
plt.scatter(x,y,color='red')
plt.title('x vs y')
plt.xlabel('x data')
plt.ylabel('y data')
plt.show()
```

```python
def predict(x,m,c):
    xx = []
    for i in x:
        xx.append(m*i+c)
    return xx
```

```python
m = 1
c = 20
yp = predict(x,m,c)
print(yp)
```

```python
def dmCalculation(x,y,yp):
    sm = 0
    n = len(x)
    for i in range(n):
        sm = x[i]*(y[i]-yp[i]) + sm
    s = (-2.0/n)*sm
    return s
```

```python
dm = dmCalculation(x,y,yp)
print(dm)
```

```python
def dcCalculation(y,yp):
    sm = 0
    n = len(x)
    for i in range(n):
        sm = (y[i]-yp[i]) + sm
    s = (-2.0/n)*sm
    return s
```

```python
dc = dcCalculation(y,yp)
print(dc)
```

```python
def errorCalculaton(y,yp):
    sm = 0
    n = len(x)
    for i in range(n):
        sm = ( (y[i]-yp[i]) *  (y[i]-yp[i]) ) + sm
    s = (1.0/n)*sm
    return s
```

```python
er = errorCalculaton(y,yp)
print(er)
```

```python
err = []
m = 0
c = 0
l = 0.0001 #The learning rate
step = 50000 #The number of iteration to perform gradiant descent
for i in range(step):
    yp = predict(x,m,c)
    er = errorCalculaton(y,yp)
    err.append(er)
    dm = dmCalculation(x,y,yp)
    dc = dcCalculation(y,yp)
    m = m - (l*dm)
    c = c - (l*dc)
print(m , c)
print(err)
```

```python
#predict a new value
xt = 65
ypredict = m*xt + c
print(ypredict)
```

```python
yp = predict(x,m,c)
plt.scatter(x,y)
plt.scatter(x,yp,color='red') #regression line
plt.plot(x,yp,color='red') #regression line
plt.show()
```
