```python
import numpy as np
from math import *
from sklearn.model_selection import train_test_split
from sklearn import datasets
from collections import Counter
```

```python
spam=[
    'সার কি বেঁচে আছেন',
    'ডাবল এপ্লাস না হলে ভার্সিটির ফরম ই দেয়া হয়না এটাই কি সৃজনশীলতা',
    'যার জন্য শিক্ষককূল প্রস্তুত নন  প্রয়োজনীয় উপকরন নেই  তাকে হিড়হিড় করে টেনে মাঠে এনে ফেললেন',
    'তো এমন নড়বড়ে ব্যবস্থাটি এমন হুড়োতাড়া করে লাঞ্চ করার অর্থ কী',
    'সৃজনশীল মানে কি সার  অন্যের লেখা চুরি করে সেটার কপি করে লেখা'
]
ham=[
    'তিনি একাধারে শিক্ষক পদার্থবিদ মুক্তিযোদ্ধা মুক্তিযোদ্ধা গবেষক লেখক কলামিষ্ট ও একজন বিজ্ঞানী',
    'বাংলাদেশের গর্ব সারা বিশ্বের গর্ব',
    'সমগ্র বাংলাদেশ আপনার দিকে তাকিয়ে আছে',
    'কেননা আপনার পারফরম্যান্স এর উপর বাংলাদেশ টিমের জয় নির্ভর করছে',
    'শুভকামনা রইলো মিস্টার অলরাউন্ডার ',
    'একমাত্র তিন পরমেটে বিশ্বের এক নাম্বার অলরাউন্ডার',
    'এমন অসাধারণ ভাবেই যেন ফিরে আসতে পারো ফের ব্যাট বলের ভুবনে'
]
```

```python
spam_len=len(spam)
ham_len=len(ham)
```

```python
c=spam.copy()
c.extend(ham)
total_len=len(c)
print(c)
print(spam_len,ham_len,total_len)
```

```python
total={}
spam_c={}
ham_c={}
for i in spam:
    for j in i.split(' '):
            if j in spam_c:
                spam_c[j]+=1
            else:
                spam_c[j]=1
print(spam_c);
for i in ham:
    for j in i.split(' '):
            if j in ham_c:
                ham_c[j]+=1
            else:
                ham_c[j]=1
print(ham_c);

for i in c:
    for j in i.split(' '):
            if j in total:
                total[j]+=1
            else:
                total[j]=1
print(total,"\n")

total_sum=sum(total.values())
spam_sum=sum(spam_c.values())
ham_sum=sum(ham_c.values())
print(total_sum,spam_sum,ham_sum)
```

```python
p_spam= spam_len/total_len
p_ham= ham_len/total_len
print(p_spam,p_ham,'\n')
```

```python
spam_probability={}

for i,j in spam_c.items():
    spam_probability[i]=j/spam_sum

print(spam_probability,"\n")
```

```python
ham_probability={}

for i,j in ham_c.items():
    ham_probability[i]=j/ham_sum

print(ham_probability,"\n")
```

```python
total_probability={}

for i,j in total.items():
    total_probability[i]=j/total_sum
print(total_probability,'\n')
```

```python
Q='সার কি বেঁচে নেই'
```

```python
spam_p_up=1
spam_p_down=1
for i in Q.split(' '):
    if i in spam_probability:
        spam_p_up*=spam_probability[i]
        spam_p_down*=total_probability[i]
    else:
        spam_p_up*=0
        
spam_p_up*=p_spam
spam_p_ans=spam_p_up/spam_p_down;
print("spam-> ",spam_p_ans,"\n")
print(int(spam_p_ans*100),"%","\n")
```

```python
ham_p_up=1
ham_p_down=1
for i in Q.split(' '):
    if i in ham_probability:
        ham_p_up*=ham_probability[i]
        ham_p_down*=total_probability[i]
    else:
        ham_p_up*=0
        
        
ham_p_up*=p_ham
ham_p_ans=ham_p_up/ham_p_down;
print("ham-> ",ham_p_ans,"\n")
print(int(ham_p_ans*100),"%","\n")
```
